```python
import pandas as pd 
import numpy as np
```

```python
pd.read_csv('VariableDefinitions.csv')
```

```python
df = pd.read_csv('Train.csv')
df
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
```

```python
df['country'] = encoder.fit_transform(df['country'])
df['age_group'] = encoder.fit_transform(df['age_group'])
df['purpose'] = encoder.fit_transform(df['purpose'])
df['main_activity'] = encoder.fit_transform(df['main_activity'])
df['info_source'] = encoder.fit_transform(df['info_source'])
df['tour_arrangement'] = encoder.fit_transform(df['tour_arrangement'])
df['package_transport_tz'] = encoder.fit_transform(df['package_transport_tz'])
df['package_accomodation'] = encoder.fit_transform(df['package_accomodation'])
df['package_food'] = encoder.fit_transform(df['package_food'])
df['package_sightseeing'] = encoder.fit_transform(df['package_sightseeing'])
df['package_guided_tour'] = encoder.fit_transform(df['package_guided_tour'])
df['package_insurance'] = encoder.fit_transform(df['package_insurance'])
df['payment_mode'] = encoder.fit_transform(df['payment_mode'])
df['package_transport_int'] = encoder.fit_transform(df['package_transport_int'])
df['most_impressing'] = encoder.fit_transform(df['most_impressing'])
df['first_trip_tz'] = encoder.fit_transform(df['first_trip_tz'])

```

```python
data = df.drop(columns=['travel_with','info_source'])
```

```python
data.isnull().sum()
```

```python
data
```

```python
mean_for_female = data['total_female'].mean()
mean_for_male = data['total_male'].mean()
```

```python
data['total_female'].fillna(value=mean_for_female, inplace=True)
data['total_male'].fillna(value=mean_for_male, inplace=True)
```

```python
X = data.drop(columns=['ID','total_cost' ])
y = data['total_cost']
```

```python
from sklearn.model_selection import train_test_split
Xtrain, Xval, ytrain , yval = train_test_split(X,y, test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
from sklearn.linear_model import LinearRegression, Lasso, Ridge
lr = LinearRegression()
la = Lasso()
ri = Ridge()
```

```python
lr.fit(Xtrain, ytrain)
la.fit(Xtrain, ytrain)
ri.fit(Xtrain, ytrain)
```

```python
lrpred = lr.predict(Xval)
lapred = la.predict(Xval)
ripred = ri.predict(Xval)
```

```python
from sklearn.metrics import mean_absolute_error

lr_mae = mean_absolute_error(yval, lrpred)
la_mae = mean_absolute_error(yval, lapred)
ri_mae = mean_absolute_error(yval, ripred)
```

```python
print(f'lr_mae: {lr_mae}')
print(f'la_mae: {la_mae}')
print(f'ri_mae: {ri_mae}')
```
